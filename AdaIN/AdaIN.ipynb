# Adaptive Instance Normalization for Style Transfer
[to see paper](https://openaccess.thecvf.com/content/CVPR2021/papers/Chandran_Adaptive_Convolutions_for_Structure-Aware_Style_Transfer_CVPR_2021_paper.pdf)

```python
import numpy as np
from PIL import Image
```

```python
x = Image.open('content image.jpg')
y = Image.open('style image.jpg')
```

```python
x
```

```python
y
```

```python
x = np.asanyarray(x)
y = np.asanyarray(y)
```

```python
np.shape(x)
```

```python
np.shape(y)
```

```python
def AdaIN(x, y):
    y_std = np.std(y)
    y_mean = np.mean(y)
    instance_normalization = (x - np.mean(x)) / np.std(x)
    t = y_std * instance_normalization + y_mean
    return t
```

```python
result = AdaIN(x, y)
```

```python
PIL_image = Image.fromarray(np.uint8(result)).convert('RGB')
PIL_image
```
